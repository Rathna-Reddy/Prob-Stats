## QQ Plots
#### Simple graphical plot to test if a random varialbe is normally distributed (or any other distribution in general) or not
#### Other statistical testing are KS tests and Anderson Darling Tests.


```python
import numpy as np
import pylab
import matplotlib.pyplot as plt
import scipy.stats as stats

#to use the QQ-Plots 
#1. first we need to sort Xi's
#2. find the percentile values of the distribution random variables 
#and use those values on 1 axes
#3. On another axes you need to take the percitile values of the distribution against which we would like to compare
#4. To use percentile, first we need to sort them and then find the percentile values
#Example: Below I would like to compare if my distrubion is Gausiaan distribution
#5. Then do QQ-Plots with Gausiaan distribution percentile values on 1 axes and values from the step 2 on another axes
#6. if all/most of the points fall on the same line, then your random variable is Gaussian in this above example.
# or in general X and Y have similar distribution.

#Standard Normal variate - Gaussian Distributin with N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size = 1000)

#0 to 100th percentiles of std-normal
for i in range(0, 101):
    print(i, np.percentile(std_normal,i))
```

```python
#generate 100 samples for N(0,1)
#try with size 1000, 10000
measurements = np.random.normal(loc=0, scale=1, size=100)

# plt.xlim(-1,1)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
#try with size 1000
measurements = np.random.normal(loc=0, scale=1, size=1000)

# plt.xlim(-1,1)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```python
#try with size 10000
measurements = np.random.normal(loc=0, scale=1, size=10000)

# plt.xlim(-1,1)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

### limitation of QQ-Plots
#### From the above we can see that as the size of the distribution increases, the more gaussian it is, so if the number of observations is small, it's hard to intrepret Q-Q plots.

```python
# compare uniform distribution against Normal distribution
#try with size = 100, 1000 
measurements = np.random.uniform(low=-1, high=1, size=1000)

# plt.xlim(-4,4)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

#### From the above we can clearly see that if the distribution doesn't match, random variables points doesn't fall on the line.

```python
#lets use QQ-Plots for uniform distribution
measurements = np.random.uniform(low=-1, high=1, size=1000)

stats.probplot(measurements, dist="uniform", plot=pylab)
pylab.show()
```

### From the above we can see that QQ-Plots can be used to compare any type of distribution and if they fall on the straight line, then they belong to the same distribution otherwise they aren't and try/compare with different distributions
### The limitation is the more the points you have the more QQ plots help us to understand about the distribution.
